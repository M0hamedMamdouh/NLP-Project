```python
import torch
import nltk
from nltk.corpus import stopwords
import re
import pandas as pd
from torch.utils.data import DataLoader
from sentence_transformers import SentenceTransformer, util, InputExample, losses
import numpy as np
from datasets import load_dataset
```

```python
dataset = load_dataset("cardiffnlp/databench", name="semeval", split="dev")
```

```python
dataset = pd.DataFrame(dataset)
metadata = dataset.dropna()
```

```python
metadata
```

```python
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def preprocess_question(question):
    return question.strip().lower()
# Download stopwords (only once)
nltk.download('stopwords')

# Define stopword set
stop_words = set(stopwords.words('english'))

def preprocess_metadata_questions(questions):
    cleaned_questions = []
    for q in questions:
        q_clean = re.sub(r"[^a-zA-Z0-9\s]", "", q.lower())
        q_clean = " ".join([word for word in q_clean.split() if word not in stop_words])
        cleaned_questions.append(q_clean)
    return cleaned_questions
```

```python
metadata_questions_clean = preprocess_metadata_questions(metadata["question"].tolist())
train_examples = [InputExample(texts=[q,q], label=1) for q in metadata_questions_clean]
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
train_loss = losses.CosineSimilarityLoss(model=embedding_model)
```

```python
embedding_model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=20, warmup_steps=100)
```

```python
metadata_embeddings = embedding_model.encode(metadata_questions_clean, convert_to_tensor=True)
```

```python
def find_relevant_metadata(user_question, metadata, metadata_embeddings, threshold=0.5):
    # Preprocess user question
    user_question = preprocess_question(user_question)
    user_embedding = embedding_model.encode(user_question, convert_to_tensor=True)

    # Compute cosine similarity
    similarities = util.cos_sim(user_embedding, metadata_embeddings).squeeze()

    # Find the most similar row
    max_similarity, max_index = torch.max(similarities, dim=0)
    max_index = max_index.item()

    if max_similarity >= threshold:
        return metadata.iloc[max_index]
    else:
        print(f"No relevant metadata found for the question (similarity: {max_similarity:.2f}).")
        return None
```

```python
counter = 0
for i in metadata.index:
  user_question = metadata.loc[i, "question"]
  answer = metadata.loc[i, "answer"]
  relevant_metadata = find_relevant_metadata(user_question, metadata, metadata_embeddings)
  if relevant_metadata is not None:
      result = relevant_metadata["answer"]
      if result == answer:
        counter += 1
      else:
        print ("XXXXXX")
      print(f"model: {result}, real: {answer}")
print (f"accuracy = {counter/len(metadata)}")
```
